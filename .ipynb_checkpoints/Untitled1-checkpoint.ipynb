Dataset source: https://www.kaggle.com/shrutimechlearn/churn-modelling

## Data Preprocessing & Exploratory Data Analysis

Customer Churn, also known as customer attrition, customer turnover, or customer defection, in the loss of clients or customers.

Bank, insurance companies, streaming services companies and telcom service companies, often use customer churn analysis and customer churn rates as one of their key business metrics because the cost of retaining existing customers is far less than acquiring a new one.

This analysis focuses on the behavior of bank customers who are more likely to leave the bank (i.e. close their bank account). I want to find out the most striking behaviors of customers through Exploratory Data Analysis and later on use some of the predictive analytics techniques to determine the customers who are most likely to churn.

Even though the banking industry has one of the lowest churn rates compared to other sectors, the impact on profitability by losing a single customer is exponentially higher than most industries.

**Dataset Explanation**

This is an international Bank with millions of customers spread all around Europe mainly in, three countries, Spain, France and Germany. In the last six months the bank detected that the churn rates started to increase when compared to the average rate, so they decided to take measures. The bank decided to take a small sample of 10,000 of their customers and retrieve some information.For six months they followed the behaviour of these 10,000 customers and analysed which stayed and who left the bank.

## **Numerical features and Target Correlation**

```python
def correlation_ratio(categories, measurements):
    fcat, _ = pd.factorize(categories)
    cat_num = np.max(fcat)+1
    y_avg_array = np.zeros(cat_num)
    n_array = np.zeros(cat_num)
    for i in range(0,cat_num):
        cat_measures = measurements[np.argwhere(fcat == i).flatten()]
        n_array[i] = len(cat_measures)
        y_avg_array[i] = np.average(cat_measures)
    y_total_avg = np.sum(np.multiply(y_avg_array,n_array))/np.sum(n_array)
    numerator = np.sum(np.multiply(n_array,np.power(np.subtract(y_avg_array,y_total_avg),2)))
    denominator = np.sum(np.power(np.subtract(measurements,y_total_avg),2))
    if numerator == 0:
        eta = 0.0
    else:
        eta = np.sqrt(numerator/denominator)
    return eta
```

```python
df_cor = df[['Age','Tenure','EstimatedSalary','Balance','CreditScore']]
for i in df_cor:
    print(i)
    print(correlation_ratio(df['Exited'],df[i]))
```

**All numerical features have a weak correlation. Age is the highest and Salary is the lowest. I decided to not drop anything**

## 4. Handling Outlier
<a id="Outlier" > </a>

```python
features = ['CreditScore','Age','Balance','EstimatedSalary','Tenure']
n = 1
plt.figure(figsize=(10,10))
for feature in features:
    plt.subplot(3,3,n)
    sns.boxplot(df[feature])
    n+=1
    plt.tight_layout()
```

**From the boxplot above, Age and CreditScore have some extreme values. Let's check and handle those outliers if it's necessary.**

```python
cek = ["CreditScore", "Age"]

def cek_outlier(df, col_name):
    q1 = df[col_name].describe()['25%']
    q3 = df[col_name].describe()['75%']
    iqr = q3-q1 
    low  = q1-(1.5*iqr)
    high = q3+(1.5*iqr)
    print(f"Lower limit {col_name} : {low}, & Upper limit : {high}")
    Outlier = df[(df[col_name] < q1 - (1.5*iqr)) | (df[col_name] > q3 + (1.5 * iqr))][col_name].index
    print(df[col_name][Outlier].values)
    
for i in cek:
    cek_outlier(df, i)
```

## Another Visualization

```python
df1 = df.copy()
def cred_score(df1):
    if (df1["CreditScore"] > 300) & (df1["CreditScore"] <= 560):
        return "Very Bad"
    elif (df1["CreditScore"] > 560) & (df1["CreditScore"] <= 650):
        return "Bad" 
    elif (df1["CreditScore"] > 650) & (df1["CreditScore"] <= 700):
        return "Fair"
    elif (df1["CreditScore"] > 700) & (df1["CreditScore"] <= 750):
        return "Good" 
    elif (df1["CreditScore"] > 750) & (df1["CreditScore"] <= 850):
        return "Great" 

def cust_tenure(df1):
    if df1["Tenure"] <= 2:
        return "Tenure_0-2" # New customers
    elif (df1["Tenure"] > 2) & (df1["Tenure"] <= 5) :
        return "Tenure_3-5" #medium tenure with the bank 
    elif (df1["Tenure"] > 5) & (df1["Tenure"] <= 8) :
        return "Tenure_6-8" # long tenure with the bank
    elif df1["Tenure"] > 8 :
        return "Tenure_9-10" # very long tenure with the bank

def cust_age(df1):
    if df1["Age"] <= 30:
        return "Age_18-30" #Youthful customers
    elif ((df1["Age"]) > 30) & ((df1["Age"]) <= 40):
        return "Age_30-40" #Mid-age customters
    elif ((df1["Age"]) > 40) & ((df1["Age"]) <= 60):
        return "Age_40-60" # Older customers
    elif (df1["Age"]) > 60 :
        return "Age_gt_60" # Pensioners
    
def cust_balance(df1):
    if df1["Balance"] < 50000.00:
        return "Low Balance"
    elif ((df1["Balance"]>50000.00) & (df1["Balance"]<100000.00)):
        return "Medium Balance"
    elif ((df1["Balance"]>100000.00) & (df1["Balance"]<200000.00)):
        return "High Balance"
    elif df1["Balance"]>200000.00:
        return "Very High Balance"
    
df1["tenure_cat"] = df1.apply(lambda df1:cust_tenure(df1), axis = 1)
df1["age_cat"] = df1.apply(lambda df1:cust_age(df1), axis = 1)
df1["creditscore_cat"] = df1.apply(lambda df1:cred_score(df1), axis = 1)
df1["balance_cat"] = df1.apply(lambda df1:cust_balance(df1), axis = 1)

df1_e = df1[df1["Exited"] == 1]
```
